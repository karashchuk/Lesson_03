```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def func(k,a,b,x):
    return k*np.cos(x-a)-b
k,a,b = 1,2,1
x = np.linspace(-2*np.pi,2*np.pi,101)
plt.plot(x,func(k,a,b,x),'blue')
b=0
plt.plot(x,func(k,a,b,x),'green')
a=0
plt.plot(x,func(k,a,b,x), 'orange')
k=2
plt.plot(x,func(k,a,b,x),'red')
```
