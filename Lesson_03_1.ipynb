```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def vector_len(a):
    q=list(map(lambda i : i**2, a))
    return np.sqrt(sum(q))

x = [3,4]
print ('Вектор {}, длина вектора = {}'.format(x,vector_len(x)))

x = [10,10,5]
print ('Вектор {}, длина вектора = {}'.format(x,vector_len(x)))

x = [10,10,5,8,2,1]
print ('Вектор {}, длина вектора = {}'.format(x,vector_len(x)))
```
